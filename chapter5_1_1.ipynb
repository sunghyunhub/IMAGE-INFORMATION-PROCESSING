```python
#라이브러리 불러오기
import cv2 #컴퓨터 비전 라이브러리
import numpy as np #행렬연산
import matplotlib.pyplot as plt #시각화 라이브러리
```

```python
def CalcHistogram(img, num_bin):
    h,w = img.shape
    hist = np.zeros(num_bin)
    for i in range(0, h):
        for j in range(0, w):
            hist[img[i, j]] += 1
    return hist
```

```python
# 예제 영상 생성하기
#img = np.array([[6,6,6,7],[4,5,5,3],[2,1,1,3],[0,0,1,3]])
img = cv2.imread("images/Lenna.png", cv2.IMREAD_COLOR) #BGR
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #RGB

h,w,c = img.shape #220,200,3 3개의 채널
print(h,w,c)
plt.imshow(img)
```

```python
# RGB별 히스토그램 생성하기
num_bin = 2 ** 8 # 0~255 : 8bit

hist_red = CalcHistogram(img[:,:,0], num_bin) # : = all 
hist_green = CalcHistogram(img[:,:,1], num_bin) # : = all 
hist_blue = CalcHistogram(img[:,:,2], num_bin) # : = all 
```

```python
#matplotlib를 이용하여 결과 히스토그램 출력하기
plt.subplot(1,3,1); plt.bar(range(0,num_bin),hist_red)
plt.subplot(1,3,2); plt.bar(range(0,num_bin),hist_green)
plt.subplot(1,3,3); plt.bar(range(0,num_bin),hist_blue)
```
