```python
#라이브러리 불러오기
import cv2 #컴퓨터 비전 라이브러리
import numpy as np #행렬연산
import matplotlib.pyplot as plt #시각화 라이브러리
```

```python
#레나를 흑백으로 읽고 numpy를 사용하여 영상의 크기를 자동으로 구하기
img = cv2.imread("images/Lenna.png", cv2.IMREAD_GRAYSCALE) #흑백영상으로 읽기
h,w = img.shape
print(h,w)
```

```python
#posterizing
step = 3
result = np.zeros((h,w)) # output 
range_x = np.linspace(0,255,step+1)
range_y = np.linspace(0,255,step)

for i in range (0,h):
    for j in range (0,w):
        for k in range (1,len(range_x)):
            if img[i,j] <= range_x[k]:
                result[i,j] = np.round(range_y[k-1]) #127.5 -> 128 반올림
                break

result = np.uint8(result)
# print(result)
# print(np.unique(result)) #중복되지 않는 유니크한 아이들만 출력

plt.subplot(1,2,1); plt.imshow(img, cmap='gray')
plt.subplot(1,2,2); plt.imshow(result, cmap='gray')
```
